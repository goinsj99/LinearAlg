Now, we use the Gram-Schmidt process to find 3 corresponding vectors that are orthonormal:

- $\tilde{q_1} = a_1 = \begin{bmatrix} 0 \\ 1 \\ 1 \end{bmatrix}$

- \begin{align*}\tilde{q_2} = a_2 - (q_1 \cdot a_2)q_1 &= \begin{bmatrix} 1 \\ 0 \\ 1 \end{bmatrix} - \left(\frac{1}{\sqrt{2}} \cdot 1\right) \cdot \frac{1}{\sqrt{2}} \cdot \begin{bmatrix} 0 \\ 1 \\ 1 \end{bmatrix}

   \end{align*}
   
Next, we normalize this vector to have $q_2 = \frac{1}{\sqrt{6}} \begin{bmatrix} 2 \\ -1 \\ 1 \end{bmatrix} $
